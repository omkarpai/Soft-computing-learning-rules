## AND

```python
i=[[0,0],[0,1],[1,0],[1,1]]
w1=1
w2=1
t=2
bias=-2

def f_signum(net):
    if (net>=0):
        return 1
    else:
        return 0

for x in i:
    net = (x[0]*w1)+(x[1]*w2)+bias
    fnet = f_signum(net)
    print("Output:",fnet,"\n")
```

# Hebbian learning rule

```python
import numpy as np

def signum(x):
    if (x>=0):
        return 1;
    else:
        return -1;

LR = 1
training_input = np.array([[1,-2,1.5,0],[1,-0.5,-2,-1.5],[0,1,-1,1.5]])


synaptic_weight = np.array([1,-1,0,0.5])

for iteration in range(5):
    for i in range(0,3):
        input_layer = training_input[i]
        
        net = np.dot(input_layer,synaptic_weight)
        
        adjustment = LR*input_layer*signum(net)
        
        synaptic_weight = synaptic_weight + adjustment
print(synaptic_weight,"\n")

```

# Supervised learning

```python
import numpy as np

def sigmoid(x):
    Lambda=1
    return 1/(1+np.exp(-1*Lambda*x))

def sigmoid_derivative(x):
    return x*(1-x)

LR = 1e-1
training_input =np.array([[0,0,1],[1,0,1],[1,1,0],[0,1,0]])
training_output = np.array([0,1,1,0])

np.random.seed(1)
synaptic_weight = 2*np.random.random_sample(3)-1

for iteration in range(200000):
    for i in range(0,4):
        input_layer = training_input[i]
        
        net = np.dot(input_layer,synaptic_weight)
        
        output = sigmoid(net)
        
        error = training_output[i]-output
        
        adjustment = LR*error*input_layer*sigmoid_derivative(output)
        
        synaptic_weight = synaptic_weight + adjustment
        
        
test_input = np.array([1,0,1])
test_output = sigmoid(np.dot(test_input,synaptic_weight))
print(test_output)
```

# Perceptron learning rule

```python
import numpy as np

def signum(x):
    if(x>=0):
        return 1
    else:
        return -1

LR = 1

training_input = np.array([[2,1,1],[0,-1,-1]])
training_output = np.array([-1,1])

synaptic_weight = np.array([0,1,0])

for iteration in range(5000):
    for i in range(0,2):
        input_layer = training_input[i]
        
        net = np.dot(input_layer,synaptic_weight)
        
        output = signum(net)
        
        error = training_output[i]-output
        
        adjustment = LR*error*input_layer
        
        synaptic_weight = synaptic_weight + adjustment
print(synaptic_weight,"\n")

test_input = np.array([0,-1,-1])
test_output = signum(np.dot(test_input,synaptic_weight))
print(test_output)
```

# Delta learning rule

```python
import numpy as np

def sigmoid(x):
    Lambda = 1
    return (1-np.exp(-1*Lambda*x))/(1+np.exp(-1*Lambda*x))

def fDashNet(x):
    return 0.5*(1-(x*x))

LR=0.1
training_input = np.array([[1,-2,0,-1],[0,1.5,-0.5,-1],[-1,1,0.5,-1]])
training_output = np.array([-1,-1,1])

synaptic_weight = np.array([1,-1,0,0.5])

for iteration in range(3):
    for i in range(0,3):
        input_layer = training_input[i]
        
        net = np.dot(input_layer,synaptic_weight)
        
        output = sigmoid(net)
        
        error = training_output[i]-output
        
        adjustment = LR*error*input_layer*fDashNet(output)
        
        synaptic_weight = synaptic_weight + adjustment

print(synaptic_weight,"\n")
    

```

# Widrow Hoff learning rule

```python
import numpy as np

LR=0.5

training_input = np.array([[2,0,1],[2,-2,1],[1,-2,-1]])
training_output = np.array([-1,1,1])

synaptic_weight = np.array([1,0,1])

for iteration in range(3):
    for i in range(0,3):
        input_layer = training_input[i]
        
        net = np.dot(input_layer,synaptic_weight)
        
        output = net
        
        error = training_output[i] - output
        
        adjustment = LR*error*input_layer
        
        synaptic_weight = synaptic_weight + adjustment

print(synaptic_weight,"\n")
```

# Tipping problem

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# New Antecedent/Consequent objects hold universe variables and membership
# functions
quality = ctrl.Antecedent(np.arange(0, 11), 'quality')
service = ctrl.Antecedent(np.arange(0, 11), 'service')
tip = ctrl.Consequent(np.arange(0, 26), 'tip')

# Auto-membership function population is possible with .automf(3, 5, or 7)
quality.automf(3)
service.automf(3)
#tip.automf(3)
# Custom membership functions can be built interactively with a familiar,
# Pythonic API
tip['low'] = fuzz.trimf(tip.universe, [0, 0, 13])
tip['medium'] = fuzz.trimf(tip.universe, [0, 13, 25])
tip['high'] = fuzz.trimf(tip.universe, [13, 25, 25])
```

```python
# You can see how these look with .view()
quality.view()
```

```python
service.view()
```

```python
tip.view()
```

```python
rule1 = ctrl.Rule(quality['poor'] | service['poor'], tip['low'])
rule2 = ctrl.Rule(quality['average'] | service['average'], tip['medium'])
rule3 = ctrl.Rule(service['good'] | quality['good'], tip['high'])

rule1.view()
rule2.view()
rule3.view()
```

```python
tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
```

```python
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)
```

```python
# Pass inputs to the ControlSystem using Antecedent labels with Pythonic API
# Note: if you like passing many inputs all at once, use .inputs(dict_of_data)
tipping.input['quality'] = 9
tipping.input['service'] = 9

# Crunch the numbers
tipping.compute()
```

```python
print (tipping.output['tip'])
tip.view(sim=tipping)
```
